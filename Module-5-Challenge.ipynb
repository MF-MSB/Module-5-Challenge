```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.

athletic_sales_2020_path = Path('Resources/athletic_sales_2020.csv')
athletic_sales_2021_path = Path('Resources/athletic_sales_2021.csv')

# Read in data and index by date
athletic_sales_2020_data = pd.read_csv(athletic_sales_2020_path)
athletic_sales_2021_data = pd.read_csv(athletic_sales_2021_path)
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020_data.head()
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_data.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020_data.dtypes
```

```python
# Check the 2021 sales data types.
athletic_sales_2021_data.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
joined_data_rows = pd.concat([athletic_sales_2020_data, athletic_sales_2021_data], axis="rows", join="inner").reset_index()
joined_data_rows.head()
```

```python
# Check if any values are null.
#joined_data_rows.isnull()
null_counts = joined_data_rows.isnull().sum()
print(null_counts)
```

```python
# Check the data type of each column
joined_data_rows.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
joined_data_rows = joined_data_rows.astype({"invoice_date": "datetime64[ns]"})

#joined_data_rows.dtypes
```

```python
# Confirm that the "invoice_date" data type has been changed.
joined_data_rows.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.

number_products_sold = pd.pivot_table(joined_data_rows, 
                                 index=['region', 'state', 'city'],
                                 values='units_sold',
                                 aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"

number_products_sold = number_products_sold.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.

number_products_sold.sort_values(by=["Total_Products_Sold"], ascending=False).head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

total_sales = pd.pivot_table(joined_data_rows, 
                                 index=['region', 'state', 'city'],
                                 values='total_sales',
                                 aggfunc='sum')

# Rename the "units_sold" column to "Total_Sales"

total_sales = total_sales.rename(columns={"total_sales": "Total_Sales"})

# Show the top 5 results.

total_sales.sort_values(by=["Total_Sales"], ascending=False).head(5)
```

### 4. Determine which Retailer had the Most Sales

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.

total_sales_retailer = pd.pivot_table(joined_data_rows, 
                                 index=['retailer','region', 'state', 'city'],
                                 values='total_sales',
                                 aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"

total_sales_retailer = total_sales_retailer.rename(columns={"total_sales": "Total Sales"})

# Show the top 5 results.

total_sales_retailer.sort_values(by=["Total Sales"], ascending=False).head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.

womens_athletic_footwear_sales = pd.DataFrame(joined_data_rows.loc[(joined_data_rows["product"] == "Women's Athletic Footwear")])
womens_athletic_footwear_sales.head(50)
```

#### Using `groupby`

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.


total_sales_retailer_2 = pd.pivot_table(womens_athletic_footwear_sales, 
                                 index=['retailer','region', 'state', 'city'],
                                 values='units_sold',
                                 aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

total_sales_retailer_2 = total_sales_retailer_2.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})

# Show the top 5 results.

total_sales_retailer_2.sort_values(by=["Womens_Footwear_Units_Sold"], ascending=False).head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.

pivot_long_form = pd.pivot_table(womens_athletic_footwear_sales, index="invoice_date",values="total_sales", aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"

pivot_long_form = pivot_long_form.rename(columns={"total_sales": "Total Sales"})

# Show the table.

pivot_long_form

```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

pivot_long_form_resampled = pivot_long_form.resample('D').sum()

#Sort the resampled pivot table in ascending order on "Total Sales".

pivot_long_form_resampled.sort_values(by=["Total Sales"], ascending=False).head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.

pivot_resample = pivot_long_form.resample('W').sum()

#Sort the resampled pivot table in ascending order on "Total Sales".

pivot_resample.sort_values(by=["Total Sales"], ascending=False).head(10)
```
